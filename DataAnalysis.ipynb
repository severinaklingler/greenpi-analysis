```python
import pandas as pd
import plotly.express as px
import datetime
import pytz
```

```python
def show_data(sensor_type, title):
    fig = px.line(df[df['sensor']==sensor_type], x='when', y='value')
    fig.update_layout(
        title=title,
        xaxis_title="Date",
        yaxis_title="Sensor value"
    )
    fig.show()
```

Load data

```python
df = pd.read_pickle('raw_data.df')

# Cleanup timedates 
timezone = 'Europe/Zurich'
df['when'] = pd.to_datetime(df['when'])
df.when.dt.tz_convert(tz=timezone)
df = df[df['when'] >= datetime.datetime(2020,5,1,0,0,0,0, pytz.timezone(timezone))]
```

```python
show_data('moisture','Moisture Sensor')
show_data('weigth','Weight')
show_data('temperature','Temperature')
```
